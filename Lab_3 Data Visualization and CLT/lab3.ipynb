        Central Limit Theorem (CLT)</strong>
    </h1>
    <p>
        Central Limit Theorem (CLT) is a foundational principle in statistics.
    </p>
    <p>
        The central limit theorem states that if we take large number of samples from any population with finite mean and variance then the distribution of the sample means will follow the normal distribution regardless of the type of the original distribution. Also the mean of these sample means will be equal to the population mean and standard error(standard deviation of the sample means) will decrease with increase in sample size.
    </p>

</div>

        <strong>Binomial Distribution</strong> - Distribution for the number of successes in a fixed number of trials.<br>
        <strong>Poisson Distribution</strong> - Models the number of occurrences of an event in a fixed interval of time or space.<br>
        <strong>Exponential Distribution</strong> - Models the time between events in a Poisson process.<br>
        <strong>Hyper-geometric Distribution</strong> - Distribution of successes in a sample drawn without replacement.<br>
        <strong>Negative Binomial Distribution</strong> - Distribution for the number of trials needed before a specified number of successes occur.

```python
import numpy as np
import matplotlib.pyplot as plt

def hist(data, bins, title='Histogram', x_label='x', y_label='y'):

    plt.figure()
    plt.hist(data, bins=bins, linewidth=0.4, edgecolor="black",color='purple')
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()

rollNo = 19
mean_normal = rollNo
std_normal = (rollNo / 100) * 2

normal_data = np.random.normal(loc=mean_normal, scale=std_normal, size=1000)

hist(normal_data, 20, title='Normal Data', x_label='Values', y_label='Frequency')

scale_exp = rollNo/10
exp = np.random.exponential(scale=scale_exp, size=1000)
hist(exp, 40, title='Exponential Data', x_label='Values',
y_label='Frequency')

poisson = np.random.poisson(lam=rollNo/10, size=10000)
hist(poisson, 15, title='Poisson Data', x_label='Values',
y_label='Frequency')

hypergeometric = np.random.hypergeometric(ngood=200, nbad=rollNo,nsample=50, size=1000)
hist(hypergeometric, 15, title='hypergeometric Data',
x_label='Values', y_label='Frequency')

binomial = np.random.binomial(n=20, p=0.9, size=1000)
hist(binomial, 20, title='Binomial Data', x_label='Values',
y_label='Frequency')


negative_binomial = np.random.negative_binomial(n=rollNo//2, p=rollNo/100,
size=2000)
hist(negative_binomial, 30, title='Negative Binomial Data',
x_label='Values', y_label='Frequency')


```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_clt(functions, paramameters, title, sample_sizes, num_samples=500):
    plt.figure()
    for i, sample_size in enumerate(sample_sizes):
        sample_mean = [np.mean(functions(*paramameters, size=sample_size)) for _ in range(num_samples)]
        sns.kdeplot(sample_mean, fill=True, alpha=0.6, label=f'Sample Size = {sample_size}')
    plt.title(f'CLT {title.upper()}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.show()

sample_sizes = [10, 20, 30, 40]
roll = 34
mean_normal = roll
std_normal = roll / 10
plot_clt(np.random.normal, (mean_normal, std_normal), 'NORMAL DISTRIBUTION', sample_sizes)

n = roll
p = roll / 100
plot_clt(np.random.binomial, (n, p), 'BINOMIAL DISTRIBUTION', sample_sizes)

n = 1
p = roll / 100
plot_clt(np.random.binomial, (n, p), 'BERNOULLI DISTRIBUTION', sample_sizes)

lam = roll / 10
plot_clt(np.random.poisson, (lam,), 'POISSON DISTRIBUTION', sample_sizes)

scale_exp = roll / 10
plot_clt(np.random.exponential, (scale_exp,), 'EXPONENTIAL DISTRIBUTION', sample_sizes)

k = roll
n = roll / 10
plot_clt(np.random.hypergeometric, (k, k - n, n), 'HYPERGEOMETRIC DISTRIBUTION', sample_sizes)

r = roll // 2
p = roll / 100
plot_clt(np.random.negative_binomial, (r, p), 'NEGATIVE BINOMIAL DISTRIBUTION', sample_sizes)
```

provided along with this assignment (or visit the site Student Performance
Dataset: Academic Insights 10K) which consists of data on various
demographics and score in different subjects for about 10000 students. For the
columns “math_score”, “reading_score”, “writing_score”, “science_score” and
“total_score”, plot data distribution (histograms) for each and then for randomsample means created from each of these data distribution (CLT) increasing samplesize in steps from 10 to 100.
    </p>

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Read the dataset
df = pd.read_csv('Student_performance_10k.csv')

columns_to_clean = ['math_score', 'reading_score', 'writing_score', 'science_score', 'total_score']

for col in columns_to_clean:
    df[col] = pd.to_numeric(df[col].replace(r'[^\d]', '', regex=True), errors='coerce')
    df[col] = df[col].fillna(df[col].mean())

sample_sizes = [10, 20, 30, 40, 50, 75, 100]
num_sample = 500

def gen_sample_mean(data, sample_size, num_sample, replace=True):
    generated_samples = np.zeros(num_sample)
    for i in range(num_sample):
        sample = np.random.choice(data, size=sample_size, replace=replace)
        generated_samples[i] = np.mean(sample)
    return generated_samples

for i, col in enumerate(columns_to_clean):
    plt.figure()
    for sample_size in sample_sizes:
        sample_means = gen_sample_mean(df[col].values, sample_size, num_sample)
        sns.kdeplot(sample_means, fill=True, alpha=0.6, label=f'Sample Size = {sample_size}')
    plt.title(f"CLT in Action for {col.capitalize()}")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.legend()
    plt.show()
```
