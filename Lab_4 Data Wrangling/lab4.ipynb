    </h1>
    <p>
        Data wrangling is the process of converting raw data into a usable form. It may also be called data munging or data remediation.
    </p>
    <p>
        Data wrangling describes a series of processes designed to explore, transform, and validate raw datasets from their messy and complex forms into high-quality data. You can use your wrangled data to produce valuable insights and guide business decision.
    </p>
    <p>
        The process for data wrangling in Python requires <strong>Pandas</strong>.
    </p>
</div>

    <p>
        The data wrangling process has four broad steps::<br><br>
        <strong>Discovery</strong> - In the discovery stage, you'll essentially prepare yourself for the rest of the process. Here, you'll think about the questions you want to answer and the type of data you'll need in order to answer them. You'll also locate the data you plan to use and examine its current form in order to figure out how you'll clean, structure, and organize your data in the following stages..<br>
        <strong>Transformation</strong> - During the transformation stage, you'll act on the plan you developed during the discovery stage. This piece of the process can be broken down into four components: structuring, normalizing and denormalizing, cleaning, and enriching.<br>
        <strong>Validation</strong> - During the validation step, you essentially check the work you did during the transformation stage, verifying that your data is consistent, of sufficient quality, and secure. This step may be completed using automated processes and can require some programming skills.<br>
        <strong>Publishing</strong> - After you've finished validating your data, you're ready to publish it. When you publish data, you'll put it into whatever file format you prefer for sharing with other team members for downstream analysis purposes.<br>
    </p>
</div>

        <strong>Problem 1.</strong> On the same data that you performed wrangling on in the lab:- 
    </p>
    <p>
      Clean the STARS column to extract new Directors and Actors columns.
    </p>
    <p>
    Then obtain a rating of each actor, by averaging the rating across all the movies that they ºve played 
in. And then plot the ratings of thus obtained top 10 actors.
    </p>
</div>


```python
import pandas as pd
import matplotlib.pyplot as plt

file_name = "movies.csv"
df = pd.read_csv(file_name)

df[['Directors', 'Actors']] = df['STARS'].str.split(',', n=1, expand=True)
df['Directors'] = df['Directors'].str.replace('Director:', '').str.strip()  # Clean the 'Directors' column
df['Actors'] = df['Actors'].str.strip()  # Clean the 'Actors' column

df = df[df['Actors'].notna() & (df['Actors'] != '')]

actors_list = df['Actors'].str.split(',').explode().str.strip()

actors_df = pd.DataFrame({
    'Actor': actors_list, 
    'Rating': df.loc[actors_list.index, 'RATING'].values  
})

average_actor_ratings = actors_df.groupby('Actor')['Rating'].mean().sort_values(ascending=False)

top_10_actors = average_actor_ratings.head(10)

plt.figure(figsize=(12, 6))
top_10_actors.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Top 10 Actors by Average Rating")
plt.xlabel("Actor")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

csv_files = [
    "spotify_artist_data_2023.csv",
    "spotify_data_12_20_2023.csv",
    "spotify_features_data_2023.csv",
    "spotify_tracks_data_2023.csv",
    "spotify-albums_data_2023.csv"
]

dataframes = {}

for file in csv_files:
    df_name = file.replace('.csv', '').replace('-', '_')
    df = pd.read_csv(file)
    dataframes[df_name] = df
    print(f"Schema for {df_name}:")
    print(df.info(), "\n")

for name, df in dataframes.items():
    print(f"Cleaning {name}...")
    
    duplicate_count = df.duplicated().sum()
    if duplicate_count > 0:
        print(f"  Removing {duplicate_count} duplicate rows.")
        df.drop_duplicates(inplace=True)
    
    missing_values = df.isnull().sum()
    if missing_values.any():
        print(f"  Missing values detected:\n{missing_values}")
        print("  Filling missing values with appropriate strategies.")
        for column in df.columns:
            if df[column].dtype in ['float64', 'int64']:
                df[column].fillna(df[column].median(), inplace=True)
            else:
                df[column].fillna("Unknown", inplace=True)
    
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

    dataframes[name] = df

print("Data cleaning completed for all files.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_name = "spotify_features_data_2023.csv"
df = pd.read_csv(file_name)

numeric_df = df.select_dtypes(include=['float64', 'int64'])

statistics = numeric_df.describe().transpose() 
statistics['variance'] = numeric_df.var()  
print("Basic Statistics for each numeric feature:")
print(statistics)

correlation_matrix = numeric_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Heatmap of Features")
plt.show()

print("\nConclusions from the heatmap:")
for column in correlation_matrix.columns:
    highly_correlated = correlation_matrix[column][correlation_matrix[column] > 0.7].index.tolist()
    if len(highly_correlated) > 1:  
        print(f"- {column} is highly correlated with {', '.join(highly_correlated)}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_name = "spotify_data_12_20_2023.csv"
df = pd.read_csv(file_name)

top_10_tracks = df[['track_name', 'track_popularity']].sort_values(by='track_popularity', ascending=False).head(10)
print("Top 10 Popular Tracks:")
print(top_10_tracks)

df['primary_artist'] = df['artists'].apply(lambda x: x.split(',')[0] if isinstance(x, str) else x)
tracks_per_artist = df['primary_artist'].value_counts()
print("\nNumber of Tracks per Artist:")
print(tracks_per_artist)

average_danceability = df.groupby('primary_artist')['danceability'].mean().sort_values(ascending=False)
top_danceability_artists = average_danceability.head(10)
print("\nArtists with the Highest Average Danceability:")
print(top_danceability_artists)

genre_columns = ['genre_0', 'genre_1', 'genre_2', 'genre_3', 'genre_4']
df['all_genres'] = df[genre_columns].apply(lambda x: ','.join(x.dropna()), axis=1)
average_popularity_by_genre = df.explode('all_genres').groupby('all_genres')['track_popularity'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
average_popularity_by_genre.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Average Popularity of Each Genre")
plt.xlabel("Genre")
plt.ylabel("Average Popularity")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

df['release_year'] = pd.to_datetime(df['release_date'], errors='coerce').dt.year  
album_releases_by_year = df[df['release_year'] > 1920].groupby('release_year').size()

plt.figure(figsize=(12, 6))
album_releases_by_year.plot(kind='line', marker='o', color='orange')
plt.title("Total Number of Album Releases Each Year (After 1920)")
plt.xlabel("Year")
plt.ylabel("Number of Album Releases")
plt.grid(True)
plt.show()

df['duration_minutes'] = df['duration_ms'] / 60000
average_duration_by_year = df[df['release_year'] > 1920].groupby('release_year')['duration_minutes'].mean()

plt.figure(figsize=(12, 6))
average_duration_by_year.plot(kind='line', marker='o', color='green')
plt.title("Average Album Duration Each Year (After 1920)")
plt.xlabel("Year")
plt.ylabel("Average Duration (Minutes)")
plt.grid(True)
plt.show()
```
