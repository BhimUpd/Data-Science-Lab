```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('autos.csv')
data.head(5)
```

```python
features = data[["highway_mpg", "engine_size", "curb_weight", "horsepower"]]
features = (features - features.mean())
features.head(5)
```

```python
covMatrix = np.cov(features, rowvar=False)
eigenValues, eigenVector = np.linalg.eigh(covMatrix)
print("eigen values: ", eigenValues)
print("eigen vectors: ", eigenVector)
```

```python
sortedIndices = np.argsort(eigenValues)[::-1]
eigenValues = eigenValues[sortedIndices]
eigenVector = eigenVector[:, sortedIndices]
print("eigen values: ", eigenValues)
print("eigen vectors: ", eigenVector)
```

```python
principalComponents = eigenVector[:, :2]
featuresAfterPCA = np.dot(features, principalComponents)
x = featuresAfterPCA[:, 0]
y = featuresAfterPCA[:, 1]
plt.scatter(x, y, color = 'red')
```

```python
df1 = pd.DataFrame(features)
df2 = pd.DataFrame(featuresAfterPCA)
import seaborn as sns
sns.pairplot(df1)
plt.suptitle("Before PCA")
plt.show()
```

```python
sns.pairplot(df2)
plt.suptitle('After PCA')
plt.show()
```

```python
varianceRatio = eigenValues/np.sum(eigenValues)
plt.bar(range(1,5), varianceRatio, color = 'red')
plt.show()
print(varianceRatio)
```
