    <p>
<strong>1) First, we randomly initialize k points, called means or cluster centroids.</strong>: <br>
<strong>2) We categorize each item to its closest mean, and we update the mean’s coordinates, which are the averages of the items categorized in that cluster so far.

<strong>3) We repeat the process for a given number of iterations and at the end, we have our clusters.

```python
import pandas as pd
data = pd.read_csv("data.csv")
print(data.head(3))
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
def plot_silhouette_scores(data, k_range):
    silhouette_scores = []
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        cluster_labels = kmeans.fit_predict(data)
        score = silhouette_score(data, cluster_labels)
        silhouette_scores.append(score)
    plt.figure(figsize=(8, 5))
    plt.plot(k_range, silhouette_scores, marker='o')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Silhouette Score')
    plt.title('Silhouette Scores for Different k Values')
    plt.show()
    return silhouette_scores
k_values = range(2, 11)
silhouette_scores_initial = plot_silhouette_scores(data, k_values)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
reduced_data = pca.fit_transform(data)
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.title('PCA Explained Variance')
plt.show()
```

```python
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(data)
```

```python
silhouette_scores_reduced = plot_silhouette_scores(reduced_data, k_values)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(k_values, silhouette_scores_initial, marker='o', label='Original Data')
plt.plot(k_values, silhouette_scores_reduced, marker='x', label='Reduced Data after PCA')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Comparison of Silhouette Scores Before and After PCA operation')
plt.legend()
plt.show()
```
