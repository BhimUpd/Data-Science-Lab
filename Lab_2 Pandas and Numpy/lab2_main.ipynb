# Calculate BMI of Students

```python
df=pd.read_csv('data_studentlist.csv')
df['BmiStudents']=df['weight']/(df['height']/100)**2
df['BmiStudents'].round(3)#A new column BmiStudents is created in df
```

# Create Pivot Table with Gender in Rows, Blood Type in Column and Count as Values

```python
import pandas as pd
import numpy as np
df=pd.read_csv('data_studentlist.csv')
df.pivot_table(index='gender',columns='bloodtype',aggfunc='size')
```

# Sort Data Frome By Gender and Blood Type


```python
import pandas as pd
import numpy as np
df=pd.read_csv('data_studentlist.csv')
df.sort_values(by=['gender','bloodtype'],ascending=[True,True])
```

# Convert  Height  to  Meter


```python
import pandas as pd
import numpy as np
df=pd.read_csv('data_studentlist.csv')
df['height']=df['height']*0.01
print(df['height'])
```

# Most Common Blood Type


```python
import pandas as pd
import numpy as np
df=pd.read_csv('data_studentlist.csv')
df.agg({'bloodtype': pd.Series.mode})
```

# Display Male and Female Avg Height


```python
import pandas as pd

import numpy as np

df=pd.read_csv('data_studentlist.csv')
dfTwo = df.groupby('gender').agg({'height':['mean']})
# dfTwo = df.groupby('gender').agg({'height':np.mean})
print(dfTwo)
```

# Male to Female Proportion


```python
import pandas as pd
import numpy as np

df=pd.read_csv('data_studentlist.csv')
maleCount=df['gender'].value_counts()['M']
femaleCount=df['gender'].value_counts()['F']

print(maleCount/femaleCount)


```
