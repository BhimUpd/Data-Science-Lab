```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('insurance.csv')
data.head(10)
```

```python
from sklearn.preprocessing import OneHotEncoder
categoricalColumns = ['sex', 'smoker', 'region']
Ohe = OneHotEncoder(drop='first', sparse_output=False)
categoriesEncoded = Ohe.fit_transform(data[categoricalColumns])
```

```python
numericalColumns =['age', 'bmi', 'children']
X = np.hstack([data[numericalColumns].values, categoriesEncoded])
y= data['charges'].values
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
x= StandardScaler().fit_transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=22)
```

```python
x_train_bias  = np.c_[np.ones(x_train.shape[0]), x_train]
x_test_bias = np.c_[np.ones(x_test.shape[0]), x_test]
initial_part = np.linalg.inv(x_train_bias.T @ x_train_bias)
final_part = x_train_bias.T @ y_train
coef = initial_part @ final_part
y_pred_normal = x_test_bias @ coef
print(coef)
```

```python
plt.scatter(y_test, y_pred_normal, color = 'lightgreen')
plt.xlabel("Actual Charges")
plt.ylabel("Predicted Charges")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='blue', label='Ideal Line')
plt.legend()
plt.title(" Actual vs Predicted Charges (Using Normal Equation Method) ")
plt.show()
```

```python
def calculateR2 (y_actual, y_predicted):
  numerator = np.sum((y_actual - y_predicted)**2)
  denumerator = np.sum((y_actual- y_actual.mean())**2)
  return (1 - (numerator/denumerator))
def calculateMSE (y_actual, y_predicted):
  return np.sum((y_actual- y_predicted)**2)/len(y_actual)
```

```python
print("The value of R2 is: ", calculateR2(y_test, y_pred_normal))
print("The value of MSE is: ", calculateMSE(y_test, y_pred_normal))
```

```python
feature_names = ['bias', 'age', 'bmi', 'children', 'sex_male', 'smoker_yes', 'region_northwest', 'region_southeast', 'region_southwest']
features_with_coef = sorted(zip(feature_names, coef), key=lambda x: x[1], reverse=True)
sorted_features, sorted_coefs = zip(*features_with_coef)
plt.figure(figsize=(10, 6))
plt.bar(sorted_features, sorted_coefs, color='green')
plt.xlabel("Features", fontsize=12)
plt.ylabel("Coefficient Value", fontsize=12)
plt.title("Feature Coefficients in Descending Order of Importance", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred_sklearn = lr.predict(x_test)
plt.scatter(y_test, y_pred_sklearn, color = 'orange')
plt.xlabel("Actual Charges")
plt.ylabel("Predicted Charges")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='blue', label='Ideal Line')
plt.legend()
plt.title(" Actual vs Predicted Charges (Using Sk learn) ")
plt.show()
```

```python
print("The value of R2 is: ", calculateR2(y_test, y_pred_sklearn))
print("The value of MSE is: ", calculateMSE(y_test, y_pred_sklearn))
```

```python
coefficients = lr.coef_
intercept = lr.intercept_
feature_names = ['bias'] + ['age', 'bmi', 'children', 'sex_male', 'smoker_yes', 'region_northwest', 'region_southeast', 'region_southwest']
coefficients_with_bias = [intercept] + list(coefficients)
features_with_coef = zip(feature_names, coefficients_with_bias)
sorted_features = sorted(features_with_coef, key=lambda x: abs(x[1]), reverse=True)
sorted_feature_names = [feature for feature, _ in sorted_features]
sorted_coefficients = [coef for _, coef in sorted_features]
plt.figure(figsize=(10, 6))
plt.bar(sorted_feature_names, sorted_coefficients, color='green')
plt.xlabel("Features", fontsize=12)
plt.ylabel("Coefficient Value", fontsize=12)
plt.title("Feature Coefficients (Using Sklearn Linear Regression)", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
alpha = 0.01
iterations = 1000
m, n = x_train.shape
coefs = np.zeros(n + 1)
for _ in range(iterations):
    gradient = (1 / m) * x_train_bias.T @ (x_train_bias @ coefs - y_train)
    coefs -= alpha * gradient
y_pred_gd = x_test_bias @ coefs
```

```python
plt.scatter(y_test, y_pred_gd, color = 'green')
plt.xlabel("Actual Charges")
plt.ylabel("Predicted Charges")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='blue', label='Ideal Line')
plt.legend()
plt.title(" Actual vs Predicted Charges (Using Gradient Descent method) ")
plt.show()
```

```python
print("The value of R2 is: ", calculateR2(y_test, y_pred_gd))
print("The value of MSE is: ", calculateMSE(y_test, y_pred_gd))
```

```python
data = pd.read_csv('points.csv')
x = data['x'].values
y = data['y'].values
plt.scatter(x, y, color = 'blue')
plt.show()
```

```python
def func(x):
    return (x - 6)**2 + np.sin(22 * x)
def gradOfFunc(x):
    return 2 * (x - 6) + 22 * np.cos(22 * x)
def gradientDescent(learning_rate, num_iterations, x_init):
    x = x_init
    history = [(x, func(x))]
    for i in range(num_iterations):
        grad = gradOfFunc(x)
        x = x - learning_rate * grad
        if i % 10 == 0:
            history.append((x, func(x)))
    return x, history
```

```python

learning_rate = 0.1
num_iterations = 200
x_init = 0
min_x, history = gradientDescent(learning_rate, num_iterations, x_init)
x_values = [h[0] for h in history]
y_values = [h[1] for h in history]
```

```python
x_range = np.linspace(-10, 20, 400)
y_range = func(x_range)
plt.plot(x_range, y_range, label='Function y = (x - 6)^2 + sin(Roll_No * x)', color='blue')
plt.scatter(x_values, y_values, color='red', label='Gradient Descent Steps', zorder=5)
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Gradient Descent to Minimize Function with roll no 22, Learning Rate: {learning_rate}')
plt.legend()
plt.show()
print(f"Minimum value of x: {min_x}")
print(f"Minimum value of y: {func(min_x)}")
```

```python
def model(x, a, b, c):
    return a * x**2 + b * x + c
def mse_loss(x, y, a, b, c):
    m = len(x)
    y_pred = model(x, a, b, c)
    return (1/m) * np.sum((y_pred - y)**2)
def gradients(x, y, a, b, c):
    m = len(x)
    y_pred = model(x, a, b, c)
    grad_a = (2/m) * np.sum((y_pred - y) * x**2)
    grad_b = (2/m) * np.sum((y_pred - y) * x)
    grad_c = (2/m) * np.sum(y_pred - y)
    return grad_a, grad_b, grad_c
def gradient_descent(x, y, learning_rate, num_iterations):
    a, b, c = 0, 0, 0
    history = []
    for _ in range(num_iterations):
        grad_a, grad_b, grad_c = gradients(x, y, a, b, c)
        a -= learning_rate * grad_a
        b -= learning_rate * grad_b
        c -= learning_rate * grad_c
        history.append((a, b, c, mse_loss(x, y, a, b, c)))
    return a, b, c, history
learning_rate = 0.0001
num_iterations = 1000
a_opt, b_opt, c_opt, history = gradient_descent(x, y, learning_rate, num_iterations)
x_range = np.linspace(min(x), max(x), 100)
y_fitted = model(x_range, a_opt, b_opt, c_opt)
plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x_range, y_fitted, color='blue', label='Fitted Parabola')
plt.xlabel('x')
plt.ylabel('y')
plt.title('The Best Fitting Parabola using Gradient Descent')
plt.legend()
plt.show()
print(f"The Optimized coefficients are a = {a_opt:.4f}, b = {b_opt:.4f}, c = {c_opt:.4f}")
```
